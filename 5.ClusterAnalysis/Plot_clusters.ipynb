```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import for 3D plotting
import pandas as pd 
```

```python
df = pd.read_csv('GoodClusters.csv')  # Replace 'your_data.csv' with your actual file path

# Separate data by class label
class1_data = df[df['Cluster Label'] == -1]
class2_data = df[df['Cluster Label'] == 0]
class3_data = df[df['Cluster Label'] == 1]
```

```python
print(len(class1_data))
print(len(class2_data))
print(len(class3_data))
```

```python
set(df['Cluster Label'])
```

```python

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for each class with different colors
ax.scatter(class1_data['dim0'], class1_data['dim1'], class1_data['dim2'], c='r', label='Cluster 1')
ax.scatter(class2_data['dim0'], class2_data['dim1'], class2_data['dim2'], c='g', label='Cluster 2')
ax.scatter(class3_data['dim0'], class3_data['dim1'], class3_data['dim2'], c='b', label='Cluster 3')

# Set labels for the axes
ax.set_xlabel('Dimension 1')
ax.set_ylabel('Dimension 2')
ax.set_zlabel('Dimension 3')

# Add a legend
ax.legend()

# Show the 3D plot
plt.show()
```
